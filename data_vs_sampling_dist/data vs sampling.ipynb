## Generation of Data
To vividly illustrate the relationship between Data Distribution and Sampling Distribution, we shall begin by generating random skewed data that will result in a non-normal (non-Gaussian) distribution. The rationale behind selecting non-normal data is to provide a more pronounced contrast that underscores the interplay between these two critical distributions.

We will now proceed to import the necessary Python plotting libraries and generate a dataset characterized by right-skewed distribution.

```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import resample # If bootstrapping is to be performed
```

```python
# Setting a random seed for reproducibility
np.random.seed(42)

# Generating right-skewed data using the exponential distribution
data = np.random.exponential(scale=2, size=1000)

# Plotting a histogram to visualize the distribution
plt.hist(data, bins=20, edgecolor='black')
plt.title('Right-Skewed Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

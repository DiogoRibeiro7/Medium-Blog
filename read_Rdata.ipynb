```python
"https://cran.r-project.org/web/packages/changepoint.np/index.html"


import pyreadr
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://github.com/cran/changepoint.np/blob/master/data/HeartRate.RData?raw=true"
dst_path = "HeartRate.RData"
res = pyreadr.read_r(pyreadr.download_file(url, dst_path), dst_path)
```

```python
data_ordered_dict = res['HeartRate']
```

```python
data_ordered_dict
```

```python
data_frame = pd.DataFrame(data_ordered_dict)
```

```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the Excel file
data = pd.read_excel("PORDATA_N.-de-dias-com-onda-de-calor.xlsx")

# Clean the data
data_clean = data.iloc[6:].copy()  # Start from 7th row
data_clean.columns = data_clean.iloc[0]  # Set the header to the 7th row
data_clean = data_clean.iloc[1:]  # Drop the 7th row since it's now the header

# Remove any trailing columns with all NaN values
data_clean = data_clean.dropna(axis=1, how='all')

# Correct the column names and make sure 'Anos' is the first column
data_clean.columns = pd.Series(data_clean.columns).fillna('Station' + pd.Series(range(1, data_clean.shape[1]+1)).astype(str))
data_clean = data_clean.rename(columns={'Station1': 'Anos'})

# Remove rows where 'Anos' is not a number
data_clean = data_clean[pd.to_numeric(data_clean['Anos'], errors='coerce').notna()]

# Convert the data to correct types
data_clean['Anos'] = data_clean['Anos'].astype(int)
for col in data_clean.columns[1:]:
    data_clean[col] = pd.to_numeric(data_clean[col], errors='coerce')

# Aggregate all columns into one
data_clean['Total'] = data_clean.iloc[:, 1:].sum(axis=1)

# Plot the data longitudinally for each station
plt.figure(figsize=(15, 10))
for col in data_clean.columns[1:9]:  # -1 to exclude the 'Total' column
    plt.plot(data_clean['Anos'], data_clean[col], label=col)
    median = data_clean[col].median()
    data_clean[col].fillna(median, inplace=True)
plt.xlabel('Year')
plt.ylabel('Number of Heatwave Days')
plt.title('Number of Heatwave Days Over Time by Weather Station')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()

# Plot histograms for the first 5 stations
stations = data_clean.columns[1:9]
for station in stations:
    plt.figure(figsize=(10, 5))
    plt.hist(data_clean[station].dropna(), bins=20)
    plt.title(f'Histogram of Heatwave Days for {station}')
    plt.xlabel('Number of Heatwave Days')
    plt.ylabel('Number of Years')
    plt.show()

    
# Plot the aggregate data over time
plt.figure(figsize=(10, 5))
plt.plot(data_clean['Anos'], data_clean['Total'])
plt.xlabel('Year')
plt.ylabel('Total Number of Heatwave Days')
plt.title('Total Number of Heatwave Days Over Time (All Stations)')
plt.show()


# Plot a histogram of the aggregate data
plt.figure(figsize=(10, 5))
plt.hist(data_clean['Total'], bins=20)
plt.title('Histogram of Total Heatwave Days (All Stations)')
plt.xlabel('Total Number of Heatwave Days')
plt.ylabel('Number of Years')
plt.show()
```

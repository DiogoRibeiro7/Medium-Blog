```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for heteroscedasticity and homoscedasticity
np.random.seed(42)  # For reproducibility
x = np.linspace(0, 10, 100)
hetero_y = 2 * x + np.random.normal(0, x)
homosced_y = 2 * x + np.random.normal(0, 3)

# Create scatter plots
plt.figure(figsize=(12, 6))

# Heteroscedasticity data
plt.subplot(1, 2, 1)
plt.scatter(x, hetero_y, color='b', alpha=0.8, label='Heteroscedasticity Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Heteroscedasticity Data')
plt.legend()

# Homoscedasticity data
plt.subplot(1, 2, 2)
plt.scatter(x, homosced_y, color='r', alpha=0.8, label='Homoscedasticity Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Homoscedasticity Data')
plt.legend()

plt.tight_layout()
plt.show()
```

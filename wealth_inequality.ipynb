These models can help us understand how wealth is distributed across different strata of society and how various factors contribute to widening or narrowing this distribution over time.

One of the most well-known models that deals with economic inequality is the Gini coefficient, a measure that captures income or wealth distribution within a population. While the Gini coefficient itself is not modeled using differential or difference equations, these equations can help us understand the factors that influence changes in the Gini coefficient over time.

Let's consider a simplified model to explore the dynamics of wealth accumulation among different income groups. Suppose we have two groups in a society: the rich and the poor. Their wealth Wr and Wp can be described by differential equations:
$$\frac{dW_r}{dt} = r_r W_r + I_r - E_r$$

$$\frac{dW_p}{dt} = r_p W_p + I_p - E_p$$
Here, rr and rp are the rates of return on investment for the rich and poor, respectively. 
Ir and Ip represent their respective incomes, and Er and Ep are their expenditures.

```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt

# Define a function to simulate wealth accumulation for two groups: rich and poor
def simulate_wealth_accumulation(T, r_r, r_p, I_r, I_p, E_r, E_p, W_r0, W_p0):
    """
    Simulates the wealth accumulation for rich and poor groups over time T based on the differential equations.
    
    Parameters:
        T (int): Total time steps
        r_r (float): Rate of return for rich
        r_p (float): Rate of return for poor
        I_r (float): Income for rich
        I_p (float): Income for poor
        E_r (float): Expenditure for rich
        E_p (float): Expenditure for poor
        W_r0 (float): Initial wealth for rich
        W_p0 (float): Initial wealth for poor
    
    Returns:
        wealth_rich (list): Wealth of rich group at each time step
        wealth_poor (list): Wealth of poor group at each time step
    """
    
    # Initialize wealth arrays
    wealth_rich = [W_r0]
    wealth_poor = [W_p0]
    
    # Simulation loop
    for t in range(1, T):
        # Calculate the new wealth based on the previous wealth and the differential equation model
        new_wealth_rich = wealth_rich[-1] + r_r * wealth_rich[-1] + I_r - E_r
        new_wealth_poor = wealth_poor[-1] + r_p * wealth_poor[-1] + I_p - E_p
        
        # Append the new wealth values to the lists
        wealth_rich.append(new_wealth_rich)
        wealth_poor.append(new_wealth_poor)
    
    return wealth_rich, wealth_poor

# Simulation parameters
T = 100  # Total time steps
r_r = 0.05  # Rate of return for rich
r_p = 0.02  # Rate of return for poor
I_r = 1000  # Income for rich
I_p = 500   # Income for poor
E_r = 800   # Expenditure for rich
E_p = 400   # Expenditure for poor
W_r0 = 10000  # Initial wealth for rich
W_p0 = 2000   # Initial wealth for poor

# Run the simulation
wealth_rich, wealth_poor = simulate_wealth_accumulation(T, r_r, r_p, I_r, I_p, E_r, E_p, W_r0, W_p0)

# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(wealth_rich, label='Wealth of Rich', color='b')
plt.plot(wealth_poor, label='Wealth of Poor', color='r')
plt.xlabel('Time')
plt.ylabel('Wealth')
plt.title('Wealth Accumulation Over Time')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from typing import List

#

def birthday_paradox_simulation(people: int) -> float:
    """
    Simulate the Birthday Paradox for a specified number of people.

    Parameters:
    people (int): The number of people in the room.

    Returns:
    float: The estimated probability of at least two people sharing a birthday.
    """
    trials = 10000
    successful_trials = 0
    
    for _ in range(trials):
        # Generate random birthdays
        birthdays = [np.random.randint(1, 366) for _ in range(people)]
        # Check for shared birthdays
        if len(birthdays) != len(set(birthdays)):
            successful_trials += 1
    
    return successful_trials / trials

def plot_birthday_paradox(max_people: int = 100):
    """
    Plot the Birthday Paradox probability as the number of people increases.

    Parameters:
    max_people (int): The maximum number of people to consider.
    """
    people_array = np.arange(1, max_people + 1)
    probabilities = np.vectorize(birthday_paradox_simulation)(people_array)
    
    plt.figure(figsize=(10, 6))
    plt.plot(people_array, probabilities, marker='o')
    plt.title('Birthday Paradox Probability')
    plt.xlabel('Number of People')
    plt.ylabel('Probability of Shared Birthday')
    plt.grid(True)
    plt.axhline(y=0.5, color='r', linestyle='--', label='50% Probability')
    plt.legend()
    plt.show()

# Call the function to visualize the Birthday Paradox
plot_birthday_paradox()
```
